## Linear shape functions in [-1,1]

```python
# Refrence element node coordinates
x0 = [-1,-1,-1]; x1 = [ 1,-1,-1]; x2 = [-1, 1,-1]; x3 = [ 1, 1,-1]
x4 = [-1,-1, 1]; x5 = [ 1,-1, 1]; x6 = [-1, 1, 1]; x7 = [ 1, 1, 1]

def L0(x):
    return (1-x)/2
def L1(x):
    return (1+x)/2

def dL0(x):
    return -1/2
def dL1(x):
    return 1/2

def N0(x,y,z):
    return L0(x)*L0(y)*L0(z)
def N1(x,y,z):
    return L1(x)*L0(y)*L0(z)
def N2(x,y,z):
    return L0(x)*L1(y)*L0(z)
def N3(x,y,z):
    return L1(x)*L1(y)*L0(z)
def N4(x,y,z):
    return L0(x)*L0(y)*L1(z)
def N5(x,y,z):
    return L1(x)*L0(y)*L1(z)
def N6(x,y,z):
    return L0(x)*L1(y)*L1(z)
def N7(x,y,z):
    return L1(x)*L1(y)*L1(z)

def dN0(x,y,z):
    return [dL0(x)*L0(y)*L0(z), L0(x)*dL0(y)*L0(z), L0(x)*L0(y)*dL0(z)]
def dN1(x,y,z):
    return [dL1(x)*L0(y)*L0(z), L1(x)*dL0(y)*L0(z), L1(x)*L0(y)*dL0(z)]
def dN2(x,y,z):
    return [dL0(x)*L1(y)*L0(z), L0(x)*dL1(y)*L0(z), L0(x)*L1(y)*dL0(z)]
def dN3(x,y,z):
    return [dL1(x)*L1(y)*L0(z), L1(x)*dL1(y)*L0(z), L1(x)*L1(y)*dL0(z)]
def dN4(x,y,z):
    return [dL0(x)*L0(y)*L1(z), L0(x)*dL0(y)*L1(z), L0(x)*L0(y)*dL1(z)]
def dN5(x,y,z):
    return [dL1(x)*L0(y)*L1(z), L1(x)*dL0(y)*L1(z), L1(x)*L0(y)*dL1(z)]
def dN6(x,y,z):
    return [dL0(x)*L1(y)*L1(z), L0(x)*dL1(y)*L1(z), L0(x)*L1(y)*dL1(z)]
def dN7(x,y,z):
    return [dL1(x)*L1(y)*L1(z), L1(x)*dL1(y)*L1(z), L1(x)*L1(y)*dL1(z)]


def dN(i, x,y,z) : 
    if i == 0:
        return dN0(x,y,z)
    elif i == 1:
        return dN1(x,y,z)
    elif i == 2:
        return dN2(x,y,z)
    elif i == 3:
        return dN3(x,y,z)
    elif i == 4:
        return dN4(x,y,z)
    elif i == 5:
        return dN5(x,y,z)
    elif i == 6:
        return dN6(x,y,z)
    elif i == 7:
        return dN7(x,y,z)
```

## Jacobian

```python
X0 = np.array([0.0, 4.0, 5.0])
X1 = np.array([0.0, 0.0, 5.0])
X2 = np.array([0.0, 4.0, 0.0])
X3 = np.array([0.0, 0.0, 0.0])
X4 = np.array([6.0, 4.0, 5.0])
X5 = np.array([6.0, 0.0, 5.0])
X6 = np.array([6.0, 4.0, 0.0])
X7 = np.array([6.0, 0.0, 0.0])

coord_array = np.array([X0,X1,X2,X3,X4,X5,X6,X7])

for i in range(8):
        display(coord_array[i])

def J(x,y,z):
    J = np.zeros((3,3))
    for k in range(8):
          for i in range(3):
              for j in range(3):
                  J[i,j] += coord_array[k][i]*dN(k,x,y,z)[j]

    return J

display(J(-0.57735, -0.57735,-0.57735))

def detJ(x,y,z):
    return np.linalg.det(J(x,y,z))

display(detJ(-0.57735, -0.57735,-0.57735))
```

## $BtCB(x)$

```python
def BtCB(x,y,z):
    return np.matmul(np.transpose(B(x,y,z)),np.matmul(C,B(x,y,z)))

def BtCBdetJ(x,y,z):
    return BtCB(x,y,z)*detJ(x,y,z)

#display(BtCB    (-0.57735,-0.57735,-0.57735))
display(BtCBdetJ(-0.57735,-0.57735,-0.57735))
```

## Gauss Quadrature Testing

```python
w0 = 1;
w1 = 1;
w2 = 1;

def gauss3D(f):
    xi_0 = [-1/np.sqrt(3), 1/np.sqrt(3)]
    xi_1 = [-1/np.sqrt(3), 1/np.sqrt(3)]
    xi_2 = [-1/np.sqrt(3), 1/np.sqrt(3)]
    
    A = 0
    for k in range(2):
        for j in range(2):
            for i in range(2):
                A += w0*w1*w2*f(xi_0[i],xi_1[j],xi_2[k])
                
    return A

 
K = gauss3D(BtCBdetJ)

display(K) 

# Eigenvalues of K
eigvals = np.linalg.eigvals(K)
display(eigvals)

```

## Displacement Testing
